```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid")

sales = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "revenue": [1200, 1350, 1280, 1500, 1650, 1720],
})

plt.figure(figsize=(8, 4))
plt.plot(sales["month"], sales["revenue"], marker="o")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
students = pd.DataFrame({
    "group": ["A", "A", "B", "B", "C", "C"],
    "score": [70, 75, 80, 85, 90, 95]
})

plt.figure(figsize=(6, 4))
sns.boxplot(data=students, x="group", y="score")
plt.title("Score Distribution by Group")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 100 random scores between 50 and 100
scores = np.random.randint(50, 101, 100)

# Step 2: Create the histogram
plt.hist(scores, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.show()

# Step 3: Explanation (printed)
print("Explanation:")
print("- Each bar shows how many scores fall into each score range (bin).")
print("- Height of the bar = number of students in that score range.")
print("- This histogram helps us see the distribution of scores, clusters, gaps, or outliers.")
```
