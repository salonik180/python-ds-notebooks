## Learning goals

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
print("Mean:", arr.mean())
print("Std:", arr.std())
```

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix":)
print("Column sums:", matrix.sum(axis=0))
print("Row sums:", matrix.sum(axis=1))
```

```python
import numpy as np

# Generate 100 random numbers between 0 and 1
numbers = np.random.rand(100)

# Compute statistics
minimum = np.min(numbers)
maximum = np.max(numbers)
mean_value = np.mean(numbers)

print("Minimum:", minimum)
print("Maximum:", maximum)
print("Mean:", mean_value)
```
