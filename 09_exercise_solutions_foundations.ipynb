## Solution: is_even(n)

```python
def is_even(n):
    return n % 2 == 0

for value in [1, 2, 15, 20]:
    print(value, is_even(value))
```

## Solution: safe_divide(a, b)

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"

print(safe_divide(10, 2))
print(safe_divide(10, 0))
```
